## **Data Cleaning**
We need to perform some data cleaning before we can start analyzing the dataset. This involves dropping irrelevant columns, renaming columns, handling missing values, and converting data types:

```python
# removing duplicates in the dataset

aid_workers_df.drop_duplicates(inplace=True)
```

```python
# Drop the first row of the dataset

aid_workers_df = aid_workers_df.drop(0)
```

```python
# Drop the first row of the dataset

aid_workers_df.drop(aid_workers_df.tail(1).index, inplace=True)
```

```python
# Drop irrelevant columns

aid_workers_df.drop(columns=['Time Precision','Geo Precision', 'Admin 1','Event Description', 'Actor Name','Programme Focus', 'Known Kidnapping or Arrest Outcome', 'SiND Event ID'], inplace=True)
```

```python
aid_workers_df.columns
```

```python
# Rename columns for clarity

aid_workers_df.rename(columns={'Country ISO': 'ISO','Location of Incident': 'Location','Weapon Carried/Used': 'Weapon','Organisation Affected': 'Organisation', 'Aid Workers Killed': 'Killed', 'Aid Workers Injured' : 'Injured', 'Aid Workers Kidnapped' : 'Kidnapped', 'Aid Workers Arrested' : 'Arrested', 'Aid Workers Killed in Captivity' : 'Killed_in_Captivity', 'International Aid Workers Killed' : 'Int\'l_Killed', 'International Aid Workers Killed in Captivity' : 'Int\'l_Killed_Captive', 'National Aid Workers Killed' : 'Nat\'l Killed', 'National Aid Workers Killed in Captivity' : 'Nat\'l_Killed_Captive', 'Female Aid Workers Killed' : 'Female_Killed', 'Female Aid Workers Killed in Captivity' : 'Female_Killed_Captive', 'Male Aid Workers Killed' : 'Male_Killed', 'Male Aid Workers Killed in Captivity' : 'Male_Killed_Captive', 'Aid Workers Injured.1' : 'Injured.1', 'International Aid Workers Injured' : 'Int\'l_Injured', 'National Aid Workers Injured' : 'Nat\'l_Injured', 'Female Aid Workers Kidnapped' : 'Female_Injured', 'Male Aid Workers Injured' : 'Male_Injured', 'Aid Workers Kidnapped.1' : 'Kidnapped.1', 'International Aid Workers Kidnapped' : 'Int\'l_Kidnapped', 'National Aid Workers Kidnapped' : 'Nat\'l_Kidnapped', 'Female Aid Workers Kidnapped' : 'Female_Kidnapped', 'Male Aid Workers Kidnapped' : 'Male_Kidnapped', 'Aid Workers Arrested.1' : 'Arrested.1', 'International Aid Workers Arrested' : 'Int\'l_Arrested', 'National Aid Workers Arrested' : 'Nat\'l_Arrested', 'Female Aid Workers Arrested' : 'Female_Arrested', 'Male Aid Workers Arrested' : 'Male_Arrested'}, inplace=True)
```

```python
# Remove rows with missing values

aid_workers_df = aid_workers_df.dropna()
```

```python
# Replacing missing values in the columns with 'Unknown'

aid_workers_df = aid_workers_df.fillna(value='Unknown')

aid_workers_df
```

## **WORKING WITH DATES**
While we've looked at overall staff member being killed, injured, kidnapped or arrested etc., it would also be useful to study these numbers on a month-by-month basis. The date column might come in handy here, as Pandas provides many utilities for working with dates.

```python
# Convert the 'Date' column to datetime format

aid_workers_df.Date
```

```python
aid_workers_df['year'] = pd.DatetimeIndex(aid_workers_df.Date).year
aid_workers_df['month'] = pd.DatetimeIndex(aid_workers_df.Date).month
aid_workers_df['day'] = pd.DatetimeIndex(aid_workers_df.Date).day
```
